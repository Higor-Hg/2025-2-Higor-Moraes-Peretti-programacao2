```python
import pandas as pd

#Criando uma série simples: 
dados = [10, 20, 30, 40, 50]

#Série indices personalizados:
vendas = pd.Series([200, 300, 400, 500], index=['Janeiro,', 'Fevereiro', 'Março', 'Abril'])
```

```python
#Ìndice personalizdo rótulo
print(vendas['Março'])

#Ìndice numérico
print(vendas[0])
```

```python
print(vendas.sum()) #Soma elementos numéricos da série;

print(vendas.mean()) #Calcule a média dos elementos númericos da série;

print(vendas.max()) #encontra o máximo elementos de uma série;
```

```python
# Pandas- Dataframe
# Estrutura bidimensionais (linhas e coluna)
# Equivalente a uma tabela ou planilha no Exel

# - Ìndices (linhas)
# - Colunas nomeadas
# - Dados heteregêneos (textos, números, datas...).

# Dados em formato de diciónario

dados = {
    'Produto': ['Arroz', 'Feijão', 'Macarrão', 'Açúcar'],
    'Preço': [5.5, 7.0, 4.5, 3.8 ],
    'Estoque': [50, 70, 40, 80]
}

#Criando o dataframe
df = pd.DataFrame(dados)
```

```python
print(df['Produto']) # Nome da coluna "Produto"

print(df[['Produto','Preço']]) # Nome das colunas: "Produto" e "Preço"

print(df.loc[0]) # Função loc[rótulo da linha]: 0

print(df.iloc[2]) # Função iloc[ídique numérico da linha]: 2 nesse caso representa a terceira coluna
```

```python
print(df.describe()) # Estatística descritivas

'''
Para dados numéricos
count: Número de valores não nulos.
mean: Média dos valores.
std: Desvio padrão dos valores.
min: Valor mínimo.
25%: Primeiro quartil (25º percentil).
50%: Mediana (50º percentil).
75%: Terceiro quartil (75º percentil).
max: Valor máximo. 
Para dados não numéricos (categóricos/objetos)
count: Número de valores não nulos.
unique: Número de valores distintos.
top: O valor que aparece com mais frequência.
freq: A frequência (contagem) do valor mais frequente. 
Exemplos de uso
df.describe(): Retorna o resumo estatístico das colunas numéricas do DataFrame df.
df.describe(include=['object']): Retorna o resumo estatístico apenas para as colunas de tipo objeto (texto).
df['coluna'].describe(): Retorna o resumo estatístico de uma coluna específica. 
'''
```

```python
print(df['Preço'].mean()) # Média dos preços da coluna "Preço"
```

```python
print(df['Estoque'].sum) # Soma dos valores da coluna
```
